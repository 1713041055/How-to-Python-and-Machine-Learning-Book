## 使用 CART 算法分类

本节我们继续请出我们熟悉的 Iris 数据集，用 CART 算法进行分类。与之前的 K 近邻不同的是，我们可以把模型的结构输出出来，直接观察每一步的判断规则。

注意，要使得决策树的结构可以直接以图片形式输出在界面上，需要首先安装 Graphviz ( 贝尔实验室出品的一个开源作图工具)，并且将其添加到系统路径。下载地址：http://www.graphviz.org/pub/graphviz/stable/windows/graphviz-2.38.msi

以下代码需在 Jupyter 中运行。


```python
# 导入需要的库。

from sklearn import tree
from sklearn.datasets import load_iris
from IPython.display import Image
import pydotplus

# 载入数据集。

iris = load_iris()
```

建立决策树模型。参数 criterion 表示最优特征的选择方法，这里是 Gini 指数。其他可调的参数还包括：树的最大深度，叶结点允许的最少样本数，内部结点允许的最少样本数等。


```python
clf = tree.DecisionTreeClassifier(criterion='gini')
clf = clf.fit(iris.data, iris.target)

# 将模型结构保存下来。

dot_data = tree.export_graphviz(clf, out_file=None)

# 显示模型结构，如下图所示。

graph = pydotplus.graphviz.graph_from_dot_data(dot_data)
Image(graph.create_png())
```
